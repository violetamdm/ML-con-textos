## Eliminar signos de puntuación y dejar letras en minúsculas

```python
import string
string.punctuation # String.punctuation no tine los carácteres ¡¿
```

Funcion **quitar_signos_y_convertir_minusculas**: 
- Recibe una lista como parámetro: lista_a_cambiar
- Elimina carácteres de puntuación y convierte todas las letras a mayúsculas
- Devuelve una lista cambiada: lista_aux

```python
def quitar_signos_y_convertir_minusculas(lista_a_cambiar):
    puntuacion = string.punctuation + "¡¿" # creo mi string de carácteres de puntuación
    str_aux = "" # string auxiliar
    lista_aux = [] # lista auxiliar
    for j in range(0, len(lista_a_cambiar)): # Recorre la lista_a_cambiar
        for i in lista_a_cambiar[j]: # Recorre los caracteres de cada string de la lista_a_cambiar
            if i in puntuacion: # Si el carácter es un caracter que quiero eliminar lo sustituye por el crácter vacío
                i = "" 
            str_aux = str_aux + i.lower() # Forma un string convertido a minúsculas
        lista_aux.append(str_aux) # Añade un string a la lista auxiliar
        str_aux = ""
    return lista_aux
```

Cambio las columnas **preguntas** y **respueatas** del dataframe 

```python
df_crucero["preguntas"] = quitar_signos_y_convertir_minusculas(df_crucero["preguntas"].tolist())
df_crucero["respuestas"] = quitar_signos_y_convertir_minusculas(df_crucero["respuestas"].tolist())
```
